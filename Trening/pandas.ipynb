### Задача создать Series

```python
import pandas as pd


def create_medications(names, counts):
    """
    Напишите функцию create_mediactions(names, counts), которая  создает Series medications,
    индексами которой являются названия лекарств names, а значениями - их количество в поставке counts
    """
    Series_medications = pd.Series(
        data = counts,
        index = names,
        name = "Series medications"
    )
    return Series_medications
    

def get_percent(medications, name):
    get_percent_namers =create_medications(names, counts).loc[name]*100 / (sum(create_medications(names, counts).iloc[::]))
    return get_percent_namers
if __name__ == '__main__':
    names=['chlorhexidine', 'cyntomycin', 'afobazol']
    counts=[15, 18, 7]
    medications = create_medications(names, counts)
    print(get_percent(medications, 'chlorhexidine')) 
```

Чтобы совместить таблицы по ключевому столбцу с помощью метода join(), необходимо использовать ключевой столбец в «правой» таблице в качестве индекса. Это можно сделать с помощью метода set_index(). Также необходимо указать название ключа в параметре on.

```python
joined = ratings_dates.join(
    movies_df.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined)
```
